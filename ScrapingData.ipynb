```python
#import libaries
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
```

```python
# Variables
BASE_URL = "https://results.eci.gov.in/AcResultGenOct2024/statewiseU08"
FIRST_PAGE_URL = "https://results.eci.gov.in/AcResultGenOct2024/statewiseU081.htm"
DATA ={"Constituency": [], "Leading Candidate": [], "Leading Party": [], "Trailing Candidate": [], "Trailing Party": [], "Margin": []}
```

```python
# GEt the html object
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}
response = requests.get(FIRST_PAGE_URL, headers=headers)
soup = bs(response.content, "html.parser")
```

```python
tbodies = soup.find_all('tbody')
print(len(tbodies))
```

```python
tbody = tbodies[0]
tds= tbody.find_all('td')
print(len(tds))
```

```python
#Let's iterate through table data tags to find which we can use
```

```python
for i in range(50):
    print(f"{i}: {tds[i].text}")
    
```

```python
index = 0
while index < len(tds):
    print(f"Constituency: {tds[index].text}")
    print(f"Leading Candidate: {tds[index+2].text}")
    print(f"Leading Party: {tds[index+4].text}")
    print(f"Trailing Candidate: {tds[index+15].text}")
    print(f"Trailing Party: {tds[index+17].text}")
    print(f"Margin: {tds[index+28].text}")
    print("\n\n")
    index+=31
```

```python
# now we need to figure out how to click the buttons to move to the next page
# Let's use selenium for that
for i in range(1,6):
    url = BASE_URL+str(i)+".htm"
    print(url)
    response = requests.get(url, headers=headers)
    soup = bs(response.content, "html.parser")
    tbodies = soup.find_all('tbody')
    tbody = tbodies[0]
    tds= tbody.find_all('td')
    index = 0
    print(f"Page:{i}")
    while index < 63:
        print(f"Constituency: {tds[index].text}")
        print(f"Leading Candidate: {tds[index+2].text}")
        print(f"Leading Party: {tds[index+4].text}")
        print(f"Trailing Candidate: {tds[index+15].text}")
        print(f"Trailing Party: {tds[index+17].text}")
        print(f"Margin: {tds[index+28].text}")
        print("\n\n")
        index+=31
```

```python
# Let's place all data in a csv file
for i in range(1,6):
    url = BASE_URL+str(i)+".htm"
    print(url)
    response = requests.get(url, headers=headers)
    soup = bs(response.content, "html.parser")
    tbodies = soup.find_all('tbody')
    tbody = tbodies[0]
    tds= tbody.find_all('td')
    index = 0
    print(f"Page:{i}")
    while index < len(tds):
        DATA["Constituency"].append(tds[index].text)
        DATA["Leading Candidate"].append(tds[index+2].text)
        DATA["Leading Party"].append(tds[index+4].text)
        DATA["Trailing Candidate"].append(tds[index+15].text)
        DATA["Trailing Party"].append(tds[index+17].text)
        DATA["Margin"].append(tds[index+28].text)
        index+=31
    print(f"Page {i} completed")
```

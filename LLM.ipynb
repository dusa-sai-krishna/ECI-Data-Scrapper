```python
import google.generativeai as genai
import pandas as pd
       
```

```python
PROMPT="""
In which district thus following assembly constituency "{}" belong to in Jammu and Kashmir.
Ouput Format: Just name of the District
"""
```

```python
with open("GEMINI_API_KEY.txt") as f:
    GEMINI_API_KEY = f.read().strip()
```

```python
genai.configure(api_key=GEMINI_API_KEY)

model = genai.GenerativeModel(model_name="gemini-1.5-flash")
response = model.generate_content("Explain how AI works")
print(response.text)
```

```python
df = pd.read_csv("districts.csv")
df
```

```python
df["District"]=None
df
```

```python
for i in range(len(df)):
    response = model.generate_content(PROMPT.format(df["Name"][i]))
    print(f"{df['Name'][i]}  - {response.text}")
    break
```
